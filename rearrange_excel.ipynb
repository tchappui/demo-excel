# Réarrangement de fichier Excel

On Commence par importer Pandas qui est la solution la plus flexible ici:

```python
import pandas as pd
```

On charge le fichier Excel et on l'analyse:

```python
xl = pd.ExcelFile('data/01-trial.xlsx')
```

```python
trials = xl.parse('Sheet1')
trials
```

La puissance des DataFrames Pandas est de pouvoir facilement être visualisé dans tous les sens:

```python
view_by_solvant = trials.sort_values(by=['Catalyst', 'Solvent'])
view_by_solvant
```

Au final, il suffit de sauver le DataFrame dans un fichier Excel:

```python
writer = pd.ExcelWriter('results.xlsx', engine='xlsxwriter')
view_by_solvant.to_excel(writer, index=False, sheet_name='Sheet1')
```
